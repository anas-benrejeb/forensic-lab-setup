```python
import hashlib
import os

def hash_file(filepath):
    md5 = hashlib.md5()
    sha256 = hashlib.sha256()
    
    with open(filepath, 'rb') as f:
        data = f.read()
        md5.update(data)
        sha256.update(data)
    
    return {
        'file': filepath,
        'md5': md5.hexdigest(),
        'sha256': sha256.hexdigest()
    }
    
```

```python
# Test on 3 files - change these paths to 3 real files on your machine
files = [
    r'C:\Users\nblenovo\Desktop\Gamedev_with_PICO-8_1.pdf',
    r'C:\Users\nblenovo\Desktop\anas.jpg', 
    r'C:\Users\nblenovo\Desktop\hello world!.txt'
]

for f in files:
    if os.path.exists(f):
        result = hash_file(f)
        print(f"FILE: {result['file']}")
        print(f"MD5:    {result['md5']}")
        print(f"SHA256: {result['sha256']}")
        print()
```

```python
# Modify the text file by adding one character
with open(r'C:\Users\nblenovo\Desktop\hello world!.txt', 'a') as f:
    f.write('x')

# Hash it again
result = hash_file(r'C:\Users\nblenovo\Desktop\hello world!.txt')
print("AFTER MODIFICATION:")
print(f"MD5:    {result['md5']}")
print(f"SHA256: {result['sha256']}")

print()
print("ORIGINAL MD5:    fc3ff98e8c6a0d3087d515c0473f8677")
print("ORIGINAL SHA256: 7509e5bda0c762d2bac7f90d758b5b2263fa01ccbc542ab5e3df163be08e6ca9")
```

## File Hashing — Forensic Integrity Tool

**What this does:** Computes MD5 and SHA-256 hashes for any file.

**Why it matters forensically:** A hash is a digital fingerprint. 
Any modification to a file — even one character — produces a 
completely different hash. This makes tampering detectable and 
is the foundation of evidence integrity in digital forensics.

**Real world use:** Investigators hash every file at seizure. 
If hashes match later, the evidence is untampered. If not, 
the chain of custody is broken.
